```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.translate.bleu_score import sentence_bleu
```

```python
df1 = pd.read_csv('imageCaption.csv')
df2= pd.read_csv('predCapt.csv')
df3= pd.read_csv('MpredCapt.csv')
```

```python
df1.head()
```

```python
df3.head(20)
```

```python
df1 = df1.sort_values('id')
df3 = df3.sort_values('id')
```

```python
df1=df1[0:13]
```

```python
df3['pred_caption1']=df3['pred_caption1'].str.upper()
df3['pred_caption2']=df3['pred_caption2'].str.upper()
df3['pred_caption3']=df3['pred_caption3'].str.upper()
df3['pred_caption4']=df3['pred_caption4'].str.upper()
```

```python
finalDf = pd.merge(df1, df3, on='id')
```

```python
finalDf
```

```python
finalDf['tokenized_caption'] = finalDf.apply(lambda row: nltk.word_tokenize(row['caption']),axis=1)
finalDf['tokenized_pred_caption1'] = finalDf.apply(lambda row: nltk.word_tokenize(row['pred_caption1']),axis=1)
finalDf['tokenized_pred_caption2'] = finalDf.apply(lambda row: nltk.word_tokenize(row['pred_caption2']),axis=1)
finalDf['tokenized_pred_caption3'] = finalDf.apply(lambda row: nltk.word_tokenize(row['pred_caption3']),axis=1)
finalDf['tokenized_pred_caption4'] = finalDf.apply(lambda row: nltk.word_tokenize(row['pred_caption4']),axis=1)
```

```python
finalDf.head(20)
```

```python
finalDf['final_pred_caption'] = finalDf.apply(lambda r: [[r['tokenized_pred_caption1']] + [r['tokenized_pred_caption2']] + [r['tokenized_pred_caption3']]+ [r['tokenized_pred_caption4']]], axis=1)
```

```python
finalDf['tokenized_caption'] = finalDf.apply(lambda r: [r['tokenized_caption']], axis=1)
```

```python
print(finalDf['tokenized_caption'][0])
print(finalDf['final_pred_caption'][0])
```

```python
finalDf['bleu_score'] =finalDf.apply(lambda row : corpus_bleu(row['final_pred_caption'],row['tokenized_caption'], weights=(1, 0, 0, 0)),axis=1)
```
